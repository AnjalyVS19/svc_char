```python
import os
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def resize_and_crop(image,size):
#     plt.imshow(image)
#     plt.show()
#     print(image.size)
    img_ratio=image.size[0]/image.size[1]
    ratio=size[0]/float(size[1])
    if ratio>img_ratio:
        image=image.resize((size[0],size[1]*int(image.size[1]/image.size[0])),Image.LANCZOS)
        image=image.crop((0,0,30,30))
    elif ratio<img_ratio:
        image=image.resize((size[0]*int(image.size[0]/image.size[1]),size[1]),Image.LANCZOS)
        image=image.crop((0,0,30,30))
    else:
        image=image.resize((size[0],size[1]),Image.LANCZOS)
    return image
```

```python
x=[]
y=[]

for path,subdir,files in os.walk(r"C:\Users\archa\OneDrive\Desktop\Downloads\English-20240510T050316Z-001\English\Img\GoodImg\Bmp"):
#     print(path)
#     print(subdir)
#     print(files)
    for filename in files:
        print(filename)
        f=os.path.join(path,filename)
        img=Image.open(f).convert('L')
        img_resizes=resize_and_crop(img,(30,30))
        print(img_resizes.size)
        img_resized=np.asarray(img_resizes.getdata(),dtype=np.float32).reshape((img_resizes.size[1]*img_resizes.size[0],1))
        x.append(img_resized)
        target=filename[3:filename.index('-')]
        y.append(target)
```

```python
x=np.array(x)
print(x.shape)
```

```python
x=x.reshape(x.shape[:2])
print(x.shape)
```

```python
classifier=SVC(verbose=0,kernel='poly',degree=3)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1)
classifier.fit(xtrain,ytrain)
prediction=classifier.predict(xtest)
print(classification_report(ytest,prediction))
```
